```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1

```python
fig1, ax1 = plt.subplots()
x = np.array([1, 5.0, 10, 15, 20])
y1 = np.array([1, 7, 3, 5, 11])
y2 = np.array([4, 3, 1, 7, 12])
ax1.plot(x, y1, label='line 1', marker='o')
ax1.plot(x, y2, label='line 2', marker='o')
ax1.legend()
plt.show()
```

## 2

```python
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([0, 7, 6, 3, 5])
y2 = np.array([9, 4, 2, 4, 9])
y3 = np.array([-7, -4, 2, -4, -7])

fig2 = plt.figure()
gs = fig2.add_gridspec(2,4)
ax1 = fig2.add_subplot(gs[0, :])
ax2 = fig2.add_subplot(gs[1, :2])
ax3 = fig2.add_subplot(gs[1, 2:])

ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)

plt.show()
```

## 3

```python
x = np.linspace(-5, 5, 100)

fig, ax = plt.subplots()
ax.plot(x, x**2)
ax.annotate('min', xy = (0, 0), xytext=(0, 10), arrowprops=dict(facecolor= 'green'))
plt.show()
```

## 4

```python
data = np.array([[np.random.randint(0,11) for i in range(0, 7)] for i in range(0, 7)])
fig4, ax4 = plt.subplots()
heatmap = ax4.imshow(data, cmap='viridis')
cbar = fig4.colorbar(heatmap, ax=ax4)
ax4.grid(False)
plt.show()
```

## 5

```python
x=np.linspace(0, 5, 1000)
plt.plot(x, np.cos(2*np.pi*x/2.5), color='red')
plt.fill_between(x, np.cos(2*np.pi*x/2.5))
plt.show()
```

## 7

```python
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([1, 2, 3, 4, 5, 6])

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 5))

for ax in (ax1, ax2, ax3):
    ax.scatter(x, y)
    ax.grid(True)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')

ax1.step(x, y, where='pre', color='blue')
ax2.step(x, y, where='post', color='green')
ax3.step(x, y, where='mid', color='purple')


plt.tight_layout()
plt.show()
```

## 8

```python
x = np.linspace(0,10, 1000)
y1 = -0.2*(x-5)**2 + 5
y2 = -0.4*(x-5)**2 + 10
y3 = -0.2*(x-10)**2 + 20
plt.plot(x, y1, color='blue')
plt.plot(x, y2, color='orange')
plt.plot(x, y3, color='green')

plt.fill_between(x, y3, color='green')
plt.fill_between(x, y2, color='orange')
plt.fill_between(x, y1, color='blue')
plt.show()
```

## 9

```python
categories = ['BMV', 'Toyota', 'Ford', 'Jaguar', 'AUDI']
values = [1.5, 1, 1, 1, 0.5]

fig10, ax = plt.subplots(figsize=(8, 6))
ax.pie(
    values,
    labels=categories,
    startangle=200,
    counterclock=False,
    explode=(0.1, 0, 0, 0, 0)
)

plt.axis('equal')
plt.show()
```

## 10

```python
categories = ['BMV', 'Toyota', 'Ford', 'Jaguar', 'AUDI']
values = [1.5, 1, 1, 1, 0.5]

fig10, ax = plt.subplots(figsize=(8, 6))
ax.pie(
    values,
    labels=categories,
    startangle=200,
    counterclock=False,
    wedgeprops={'width': 0.4, 'edgecolor': 'white'}
)

ax.add_artist(plt.Circle((0, 0), 0.3, color='white'))

plt.axis('equal')
plt.show()
```
