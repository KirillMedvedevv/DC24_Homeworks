```python
#   color
# - 'blue'
# - 'rbgcmyk' -> rg
# - RRGGBB - 'FF00EE'
# RGB (255, 255, 255)
# HTML

ax = plt.axes()
ax.plot(x, np.sin(x), color='blue', linestyle="solid")
ax.plot(x, np.sin(x-1), color='g', linestyle="dashed")
ax.plot(x, np.sin(x-2), color='0.75', linestyle="dashdot")
ax.plot(x, np.sin(x-3), color='#FF00EE', linestyle="dotted")
ax.plot(x, np.sin(x-4), color=(1, 0.2, 0.3))
ax.plot(x, np.sin(x-5), color='salmon')
plt.show()
```

```python
fig, ax = plt.subplots(4)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.sin(x))
ax[2].plot(x, np.sin(x))
ax[3].plot(x, np.sin(x))

ax[0].set_xlim(-2, 12)
ax[0].set_ylim(-1.5, 1.5)

ax[1].set_xlim(12, 24)
ax[1].set_ylim(-1.5, 1.5)

ax[3].autoscale(tight=True)
plt.show()
```

```python
# 
plt.subplot(3,1,1)
plt.plot(x, np.sin(x))
plt.title("")
plt.xlabel("x")
plt.ylabel("sin(x)")

plt.subplot(3, 1, 2)
plt.plot(x, np.cos(x))
plt.title("")
plt.xlabel("x")
plt.ylabel("cos(x)")

plt.subplot(3, 1, 3)
plt.plot(x, np.cos(x))
plt.plot(x, np.sin(x))
plt.axis('equal')
plt.title("  ")
plt.xlabel("x")
plt.ylabel("y")

plt.subplots_adjust(hspace=1)
plt.show()
```

```python
x = np.linspace(0, 10, 30)

# plt.plot(x, np.sin(x), 'o') # > s ^
plt.plot(x, np.sin(x), '--p', markersize=10, linewidth=2, markeredgecolor='red', markerfacecolor="white")
plt.show()
```

```python
x = np.linspace(0,10,30)
rng = np.random.default_rng(0)

colors = rng.random(30)
sizes = 300*rng.random(30)
plt.scatter(x, np.sin(x), marker='o', c=colors, s=sizes) # plot 
plt.colorbar()
plt.show()
```

```python
# 

x = np.linspace(0, 10, 30)
dy = 0.4
y=np.sin(x) + dy*np.random.randn(30)

plt.errorbar(x, np.sin(x), yerr=dy, fmt='.k')
plt.fill_between(x, y-dy, y+dy, color='red', alpha=0.4)

plt.show()
```

```python
def f(x, y):
    return np.sin(x)**5 + np.cos(20+x*y)*np.cos(x)

x = np.linspace(0, 5, 1000)
y = np.linspace(0, 5, 1000)
X, Y = np.meshgrid(x,y)

Z = f(X, Y)

#plt.contour(X, Y, Z, color='green')
#plt.contour(X, Y, Z, cmap="RdGy")
c = plt.contour(X, Y, Z, color = 'red')
plt.clabel(c)
plt.imshow(Z, extent=[0, 5,0,5], cmap="RdGy", interpolation='gaussian', origin='lower')
plt.colorbar()
plt.show()
```
