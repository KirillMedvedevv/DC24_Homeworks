# Машинное обучение
ML решает следующую задачу:
- Пусть имеется набор данных. Каждый элемент этого набора представим многомерным вектором. Требуется заданный набор подогнать под соответсвующую функцию, которая улавливает важные сигналы и игнорирует помехи. После создания, на новых данных функция должна работать корректно

Основные типы ML
- Обучение с учителем: Моделирует отношения между признаками и метками. Модели служат для предсказания меток на основе маркированный обучающих данных. После обучения метки присваиваются неизвестным данным
- Обучение без учителя: Моделирует признаки бе меток. Модели служат для выявления структуры немаркированных данных
- Методы частичного обучения: Не все данные промаркированы.
- Методы обучения с подкреплением(reinforcement learning): Система обучения улучшает свои характеристики на основе взаимодейсвтия со средой. При этом взаимодействии система получает сигналы(функции наград), которые несут информацию о том, насколько хорошо система решила задачу с точки зрения среды

Примеры задач на ОчУ:
- Задачи классификации
- Задачи регрессии

Примеры задач на ОбУ:
- Задача кластеризации (выделить отдельные группы данных. Группы неизвестны)
- Задача понижения размерности(поиск более сжатого представления данных)

## О структуре и терминологии

- Строки - отдельные объекты - образцы (sample)
- Столбцы - признаки - соответсвуют конкректным наблюдениям (features)
- Матрицы признаков (features matrix) размер (Число образцов x число признаков)
- Целевой массив, массив меток (targets) - одномерный массив (1 x число образцов), данные которые хотим предсказать на основе имеющихся данных
- Зависимые (метка) и независимые(признаки) переменные

## Процесс построения системы машинного обучения
1. Предварительная обработка
    - На вход поступают необработанные данные и метки
    - Происходит выбор, масштабирование признаков
    - Происходит понижение размерности(выбор сильно корелирующих, зависимых столбцов)
    - Выборка образцов
    - Получаем обучающий и тестовый наборы данных
2. Обучение модели
    - Выбор модели
    - Перекрестная проверка
    - Метрики эффективности
    - Оптимизация гиперпараметров. Это параметры, которые получаются не из данных, а являются характеристиками модели
3. Оценка и формирование финальной модели
4. Прогонозирование (использование модели)

### SciKit-learn

1. Выбираем класс модели
2. Устанавливаем гиперпараметры модели
3. Создаём матрицу и целевой массив
4. Обучение модели fit()
5. Применимости модели к новым данным
    - Predict() (учитель)
    - Transform() (Без учителя)


```python
np.random.seed(1)
x = 10*np.random.rand(50)
y= 2*x - 1 + np.random.randn(50)

from sklearn.linear_model import LinearRegression

model = LinearRegression()

X = x[:, np.newaxis]
model.fit(X, y)
```

```python
print(model.coef_[0])
print(model.intercept_)

x1 = np.linspace(0, 10, 30)
y2 = model.coef_[0]*x1 + model.intercept_

plt.plot(x1,y2)
plt.scatter(x, y)
plt.grid()
plt.show()
```

```python
#Применим к новым данным

xfit = np.linspace(-10, 10, 50)
yfit = model.predict(xfit[:, np.newaxis])

plt.plot(xfit, yfit)
plt.scatter(x, y)
```
